ML FLOW

-The most imp thing of ML Flow is : tracking server

why? -> we need to track each and every experiments.

-for that we need to create server which has mlflow functionality.

why? ->as it provides a lot of tools like comparison with other experinments and also visualise them.


```python
import mlflow
```

```python
#it ensures that whatever experiments are gonna happen here it should be tracked in this url
#and tracking server is running in our local.
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
mlflow.set_experiment("check localhost connections")

with mlflow.start_run():
    mlflow.log_metric("test",1)
    mlflow.log_metric("aayush",2)
```
