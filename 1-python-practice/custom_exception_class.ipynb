```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("enter dob"))
age = 2024-year

try:    
    if age<=30 and age>=20:
        print("age valid. you can apply.")
    else:
        raise dobException
except dobException:
    print ("sorry, your age is out of bounds")

```
