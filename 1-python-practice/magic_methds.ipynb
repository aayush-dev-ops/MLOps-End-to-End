Magic methods : Dunder methods(double underscore methods)

predefined methods in python that we can verride to change the behaviour of our objects.

like ->

```python
'''
__init__, __str__, __len__ , __dir__ etc...
'''
```

```python
#basic magic methods

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

aayush = Person("Aayush",25)
print(Person)
print(aayush)
```

```python
#basic magic methods
# common one -> __str__ -> used for string representation of an object.

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"this is nothing bt {self.name}, and i am {self.age} years old."
    def __repr__(self):
        return f"Person(name = {self.name}), age= {self.age})"
        

aayush = Person("Aayush",25)
print(Person)
print(aayush)
[print(repr(aayush))]
```
