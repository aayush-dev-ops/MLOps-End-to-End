Exception handling-
events that disrupt the normal flow of program and they occus when an error is encountered during program exectution.
 
 division by zero
 file not found error
 value error

```python
a=b
```

```python
## try catch blocks
try:
    a = b
except:
    print("hey, the variable has not been assigned")
```

```python
# using namerror as a class
try:
    a = b
except NameError as ex:
    print(ex)
```

```python
result = 1/0
```

```python
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex, "please enter smth other than zero")
```

```python
## the parent class -> Exception. it will handle all. Make sure that we areusing it at the end.
try:
    result = 1/0
    a=b
except Exception as ex:
    print(ex, ", the main exoecption got caught here.")
```

```python
'''
practical example. do not use the parent Excpetion like above in middle. Always use it a the end.
otherwise beech me wo sab ko ignore type kar deta hai.
'''
try:
    num = int(input("Enter a number: "))
    div = 10/num
    print(div)
except ValueError:
    print("input should be a number")
except ZeroDivisionError:
    print("pls enter smth other than zero")
```

```python
'''
what is happening here is that the except blckswill work when ther is an error.
and in this case the else will not be used. it will not be executed.
but if there is not  any errror then else block will be executed, right after the try block is executed.
'''

try:
    num = int(input("enter a number: "))
    result = 10/num
except ValueError:
    print("input should be a number")
except ZeroDivisionError:
    print("pls enter smth other than zero")

else:
    print(f"the result is {result}")
    
```

TRY, EXCEPT, ELSE,.... NOW - "FINALLY"

```python
# fianlly block will print at the end. whatever be the case.KUCH BHI HOOO.
try:
    num = int(input("enter a number: "))
    result = 10/num
except ValueError:
    print("input should be a number")
except ZeroDivisionError:
    print("pls enter smth other than zero")

else:
    print(f"the result is {result}")

finally:
    print("execution complete")
```
