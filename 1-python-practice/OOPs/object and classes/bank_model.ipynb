```python
class Account():
    def __init__(self,name,balance=0):
        self.name = name
        self.balance = balance

    def deposit(self, deposit):
        self.balance += deposit
        print(f"An amount of rs {deposit} has been deposited.")

    def withdraw(self, amount):
        if self.balance<amount:
            print("insufficient balance")
        else:
            self.balance -= amount
            print(f"An amount of rs {amount} has been withdrawn. New balance is {self.balance}")
    def balance_left(self):
        print(self.balance)

```

```python
holder1= Account("Aayush")
holder1.deposit(100)
holder1.withdraw(50)

```
