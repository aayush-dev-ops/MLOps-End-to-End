Methods and Attributes(instance variables).
We should create tge attributes while creating the object itslef.

```python
class Dog:
    # constructor: it is used to make sure that i will be able to define all the attribute
    # which will be avilable in this class
    # here by giving name and age, i am making sure tha i have to provide these 2 values in the object
    # during the runtime
    def __init__(self, name,age):
        self.name = name
        self.age  = age

dog1 = Dog("aayush",20)
dog2 = Dog("Raja",2)

print(dog2.name, dog2.age)
print(dog1.age, dog1.name)
```

Class with instance method. We can see that all the objects are using the same type of method. (raja and tamu says woof.)

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog("Raja", 3)
dog1.bark()

dog2 = Dog("Tamu", 7)
dog2.bark()
```
