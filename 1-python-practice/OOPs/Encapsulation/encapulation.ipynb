Abstraction -> Hiding the data.

Encapsulation -> 
Bundling together data and methods, that operate on the data within a single unit. 
Wrapping the data (variable) and metods (functions) together as a single unit. 
It restrcists direct access to some of the object's components, that prevents accidental interference and misuse of data. 

Encapsulation is done by -> Getter and Setter Methods.
Access Variables -> Public, Protected and Private Variables.


```python
# Access Variables -> Public, Protected and Private Variables.

class Person:
    def __init__(self,name,age):
        self.name = name  #public variable
        self.age = age    #public variable

def get_name(person):
        return person.name

person = Person("Aayush",24)       
get_name(person)
```

```python
dir(person) # here we can see that age and name are available and it means we can use iy anywhere(Public)
```

ENCAPSULATION WITH GETTER AND SETTERS


```python
#class k andar hmlog methods banate h as buttons(getter & setters) jo ki bahar wale use kar sakee. 

class Personal:
    def __init__(self,name ,age):
        self.__name = name  #private
        self.__age = age    #private

    # getter method for name. We areshowing the world only the button get name, he will never know what is the variable name.
    def get_name(self):
        return self.__name 
    
    #setter method for name. it will help us to change the value of a prvate variable.
    def set_name(self,name):
        self.__name = name
    
    def get_age(self):  #getter for age
        return self.__age
    
    def set_age(self,age):   #setter for age
        self.__age = age



aayush = Personal("Aayush", 25)

print(aayush.get_name())
aayush.get_age()

aayush.set_age("kittu")
print(aayush.get_age())

        
```
