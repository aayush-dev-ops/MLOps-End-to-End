Polymorphism:
method overloading - allows a child class to provide a specific implementation of a methof that is lready defined in a parent class.

for example-parent class me ek method h which says - dog barks "woof woof".
in the subclass we have to make a similar method.. kinda override saying this dog barks "bhau bhau bhau"

```python
# base class
class Animal:
    def speak(self):
        return "sound of animal"

# derived class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"

# derived class 2
class Cat(Animal):
    def speak(self):
        return "Mew!"

```

```python
raja = Dog()
raja.speak()
```

```python

golu = Cat()
golu.speak()
```

```python
#function that crates polymorphism
def animal_speak(animal):
    print(animal.speak())

raja = Dog()
print(raja.speak())

golu = Cat()
print(golu.speak())

animal_speak(raja)
```

```python
class Shape:
    def area(self):
        return "the area of a figure"
```

```python
class Square(Shape):
    def __init__(self,side):
        self.side = side
    
    def area(self):
        areaofsquare= self.side* self.side
        return f"area is {areaofsquare}"

class Rectange(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width

    def area(self):
        areaofrectangle= self.length*self.width
        return f"area of rectangle is {areaofrectangle}"
    
rec = Rectange(20,10)
rec.area()
```

```python
sq = Square(10)
sq.area()
```

INTERFACES // ABSTRACT BASE CLASSES

imp -> used to define common methods for a grp of methods.

```python
from abc import ABC,abstractclassmethod

#ABSTRACT CLASS -> EMPTY CLASSES
class Vehicle(ABC):
    @abstractclassmethod
    def start_engine(self):
        pass

#derived class
class Car(Vehicle):
    def start_engine(self):
        return "Car enigine start"
    
#derived class 2
class Bike(Vehicle):
    def start_engine(self):
        return "bike engine started"
    
#function that demonstrate polymorphism: 
def start_vehicle(Vehicle):
    print(Vehicle.start_engine())


#create objects of car and bike
bmw = Car()
tvs = Bike()

start_vehicle(bmw)
```
