Generators: Subclass of iterators. use Yeild keyword to produce a series of value laziely.
which means they generate values on the fly and do not store them in memory

```python
def square(n):
    for i in range(3):
        return i**2 # this will just take the first element and show the result based on that
```

```python
square(2)
```

```python
def squares(n):
    for i in range(3):
        yield i**2   #this will sue lazy loading and take all elements one by one
```

```python
a= squares(3)
try:
    next(a)
except StopIteration:
    print("no more elements")

next(a)
```

```python
for i in squares(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

PRACTICAL EXAMPLE OF GENERATORS:
READING LARGE FILES
